```python
first_df_reduced = survey[[ 'what_is_your_age_range','what_is_your_gender','how_often_do_you_consume_coffee_decaf_or_regular_coffee','do_you_consume_any_type_of_supplement_before_or_after_engaging_in_physical_activities']].copy()

#reducing the database by creating a copy where I will have only the columns I need
```

```python
first_df_reduced['cafe_regular'] = first_df_reduced['how_often_do_you_consume_coffee_decaf_or_regular_coffee'].apply(
    lambda x: 1 if isinstance(x, str) and x.strip().lower() in ['daily', 'weekly', 'monthly'] else 0
)
#creating a boolean only for who drinks coffe on a daily, weekly or monthly basis
```

```python
first_df_reduced['supplements'] = first_df_reduced['do_you_consume_any_type_of_supplement_before_or_after_engaging_in_physical_activities'].apply(
    lambda x: 1 if isinstance(x, str) and 'yes' in x.lower() else 0
)
#creating a boolean only for who consumes any type of supplements
```

```python
age_map = { 'Under 18 years old': 16, '18-24 years old': 21, '25-34 years old': 29.5,'35-44 years old': 39.5,'45-54 years old': 49.5,'55-64 years old': 59.5,'65 years old or older': 70}
first_df_reduced['idade_media'] = first_df_reduced['what_is_your_age_range'].map(age_map)

#as I did the survey with ranges of ages I'm trying to bring an average so I can run a comparison later from younger vs. older people. 
```

```python
col_suplementos = 'if_yes,_what_type_of_supplement_do_you_consume_check_all_that_apply'
```

```python
#trying to understand the most frequent supplements 
tipos_raw = survey[col_suplementos].dropna()
suplementos_exploded = tipos_raw.str.split(';').explode().str.strip()
suplemento_counts = suplementos_exploded.value_counts()
```

```python
suplemento_counts.head(10)
#we can see that protein is super important
```

```python
def relevant_supplements(s):
    if pd.isna(s):
        return 0
    s = s.lower()
    return int(any(x in s for x in ['protein powder', 'creatine', 'pre-workout', 'Vitamins']))

survey['relevant_supplements'] = survey[col_suplementos].apply(relevant_supplements)

col_cafe_funcional = 'have_you_ever_consumed_supplemented_coffees_coffees_enriched_with_additional_ingredients_such_as_proteins,_vitamins,_or_herbs'
survey['drink_functional_coffee'] = survey[col_cafe_funcional].apply(lambda x: 1 if isinstance(x, str) and 'yes' in x.lower() else 0)

contingencia_funcionais = pd.crosstab(survey['relevant_supplements'], survey['drink_functional_coffee'])

#test chi-square
chi2_func, p_func, _, _ = chi2_contingency(contingencia_funcionais)
p_func
```

```python
#create a boolean to check if engaging in phisycal activity influences the person to drink coffee or supplements


#if its active in phsysical activities boolean
survey['Is_active'] = survey['do_you_engage_in_any_physical_activity_regularly'].apply(
    lambda x: 1 if isinstance(x, str) and 'yes' in x.lower() else 0
)

#coffee consumption boolean
survey['Drink_coffee'] = survey['how_often_do_you_consume_coffee_decaf_or_regular_coffee'].apply(
    lambda x: 1 if isinstance(x, str) and x.strip().lower() in ['daily', 'weekly', 'monthly'] else 0
)

#Supplements consumption boolean
survey['consume_supplements'] = survey['do_you_consume_any_type_of_supplement_before_or_after_engaging_in_physical_activities'].apply(
    lambda x: 1 if isinstance(x, str) and 'yes' in x.lower() else 0
)

survey[['Is_active', 'Drink_coffee', 'consume_supplements']].head()
```

```python
contingency = pd.crosstab(survey['Is_active'], survey['Drink_coffee'])
chi2_1, p1, _, _ = chi2_contingency(contingency)

#Hipothesis: coffee vs. lifestyle

survey['Is_active_and_drink_coffee'] = ((survey['Is_active'] == 1) & (survey['Drink_coffee'] == 1)).astype(int)

contingency2 = pd.crosstab(survey['Is_active_and_drink_coffee'], survey['consume_supplements'])
chi2_2, p2, _, _ = chi2_contingency(contingency2)


{"Hip 1 – Physical activity vs coffee consumption (p-value)": round(p1, 4),"Hip2 – active lifestyle vs supplements consumption (p-value)": round(p2, 4)}
```
